```python
import pandas as pd   
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
sns.countplot(df, x='class',color = 'red' )
plt.title('Number of passengers in each class')
plt.xlabel('Class')
plt.ylabel('Number of passengers')
plt.show()
```

```python
sns.histplot(df, x="age", kde = True, color = 'purple') #kde defines wave 
plt.title('age distribution of passengers')
plt.xlabel('age')
plt.ylabel('Number of passsengers')
plt.show()
```

```python
sns.lineplot(df, x ='class', y ='fare', linestyle = 'dashed', marker ='o', color = 'green')
plt.title('Fare of each class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```
